```python
import torch
import numpy as np
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70], 
                   [74, 66, 43], 
                   [91, 87, 65], 
                   [88, 134, 59], 
                   [101, 44, 37], 
                   [68, 96, 71], 
                   [73, 66, 44], 
                   [92, 87, 64], 
                   [87, 135, 57], 
                   [103, 43, 36], 
                   [68, 97, 70]], 
                  dtype='float32')

# Targets (apples, oranges)
targets = np.array([[56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119],
                    [57, 69], 
                    [80, 102], 
                    [118, 132], 
                    [21, 38], 
                    [104, 118], 
                    [57, 69], 
                    [82, 100], 
                    [118, 134], 
                    [20, 38], 
                    [102, 120]], 
                   dtype='float32')
```

```python
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print(inputs)
print(targets)
```

```python
print(inputs.shape)
print(targets.shape)
```

```python
# Weights and Biases

w = torch.randn(3, 2, requires_grad = True)
b = torch.randn(2, requires_grad = True)
print(w)
print(b)
```

```python
def model(x):
    return torch.matmul(x,w) + b
```

```python
# Genreate Prediciton

prediction = model(inputs)
print(f"predictions: {prediction}")
```

```python
print(f"targets: {targets}")
```

```python
error = targets - prediction
print(f"error: {error}")
```

```python
# MSE
def mse_error(tar,pre):
    diff = tar - pre
    return torch.sum(diff * diff)/diff.numel()
```

```python
mse_loss = mse_error(targets,prediction)
print(f"Mean Square Error: {mse_loss}")
```

```python
mse_loss.backward()
```

```python
mse_loss
```

```python
print(f"weight: {w}")
print(f"weight.grad: {w.grad}")
```

```python
with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
```

```python
w.grad.zero_()
b.grad.zero_()
print(w.grad)
print(b.grad)
```

```python
pred2 = model(inputs)
print(pred2)
```

```python
mse_loss2 = mse_error(targets, pred2)
print(f"MSE Error after 1 backprop: {mse_loss2}")
```

```python
for i in range(1,101):
    predi = model(inputs)
    mse_lossi = mse_error(targets, predi)
    mse_lossi.backward()
    with torch.no_grad():
        w -= w.grad * 1e-5
        b -= b.grad * 1e-5
        w.grad.zero_()
        b.grad.zero_()
    print(f"Loss in epoch{i}: {mse_lossi}")
```
